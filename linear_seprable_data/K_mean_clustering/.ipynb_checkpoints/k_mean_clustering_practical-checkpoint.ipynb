```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
data=load_iris()
dataset=pd.DataFrame(data.data, columns=data.feature_names)
dataset.head()
```

```python
sns.pairplot(data=dataset)
plt.show()
# so data is linearly separable
```

## now, we will check k total kitny clusters is k ander hn, via WCSS 

```python
from sklearn.cluster import KMeans
```

```python
wcss =[]
for i in range(1,20):
    kmean=KMeans(n_clusters=i,init='k-means++',)
    kmean.fit(dataset)
    wc=kmean.inertia_  # y inertia hi basically wcss ki value h
    wcss.append(wc)
    print(wc)
```

```python
plt.figure(figsize=(10,4))
plt.plot([i for i in range (1,20)], wcss, marker="o")
plt.ylim(min(wcss) * 0.9, max(wcss) * 1.1) 
plt.xticks([i for i in range (1,20)])
plt.grid(axis= 'x',)
plt.show()
```

```python
val=KMeans(n_clusters=3, init='k-means++',)
val.fit(dataset)
```

```python
output=pd.Series(val.predict(dataset))
```

```python
dataset['target']=output
```

```python
sns.pairplot(data=dataset, hue='target')
plt.show()
```
